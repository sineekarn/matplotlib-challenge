```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_csv = "../Resources/mouse_drug_data.csv"
clinical_trial_data_csv = "../Resources/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_pd = pd.read_csv(mouse_drug_data_csv)
clinical_trial_data_pd = pd.read_csv(clinical_trial_data_csv)

# Merge two dataframes into a single dataframe using a left join
merge_table = pd.merge(clinical_trial_data_pd, mouse_drug_data_pd, on="Mouse ID", how="left")

# Display the data table for preview
merge_table.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = merge_table.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
mean_tumor_volume_df = pd.DataFrame(mean_tumor_volume)

# Reset DataFrame Index
mean_tumor_volume_df.reset_index(inplace = True)

# Preview DataFrame
mean_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error_tumor_volumes = merge_table.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
standard_error_tumor_volumes_df = pd.DataFrame(standard_error_tumor_volumes)

# Reset DataFrame Index
standard_error_tumor_volumes_df.reset_index(inplace = True)

# Preview DataFrame
standard_error_tumor_volumes_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames and use "Timepoint" as an index so they will be used as lables
mean_tumor_volume_df_pivot = mean_tumor_volume_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

# Preview that Reformatting worked
mean_tumor_volume_df_pivot.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames and use "Timepoint" as an index so they will be used as lables
standard_error_tumor_volumes_df_pivot = standard_error_tumor_volumes_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

# Preview that Reformatting worked
standard_error_tumor_volumes_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_tumor_volume_df_pivot.index, mean_tumor_volume_df_pivot["Capomulin"], yerr=standard_error_tumor_volumes_df_pivot["Capomulin"], marker="o", color="red", linestyle='--', label = "Capomulin")
plt.errorbar(mean_tumor_volume_df_pivot.index, mean_tumor_volume_df_pivot["Infubinol"], yerr=standard_error_tumor_volumes_df_pivot["Infubinol"], marker="^", color="blue", linestyle='--', label = "Infubinol")
plt.errorbar(mean_tumor_volume_df_pivot.index, mean_tumor_volume_df_pivot["Ketapril"], yerr=standard_error_tumor_volumes_df_pivot["Ketapril"], marker="s", color="green", linestyle='--', label = "Ketapril")            
plt.errorbar(mean_tumor_volume_df_pivot.index, mean_tumor_volume_df_pivot["Placebo"], yerr=standard_error_tumor_volumes_df_pivot["Placebo"], marker="d", color="black", linestyle='--', label = "Placebo")                         
 
# Set a Title and labels
plt.title("Tumor Response to Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")

# Set x & y limits of our scatter plot 
plt.xlim(-1, 50)
plt.ylim(30,80)

# Create a legend for the chart
plt.legend(loc="upper left")

#Set a grid on the plot
plt.grid()

plt.tight_layout()

# Save the Figure
plt.savefig("../Images/Tumor_Response_to_Treatment.png")

# Show the figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic_sites = merge_table.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
mean_metastatic_sites_df = pd.DataFrame(mean_metastatic_sites)

# Reset DataFrame Index
mean_metastatic_sites_df.reset_index(inplace = True)

# Preview DataFrame
mean_metastatic_sites_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_metastatic_sites_df_pivot = mean_metastatic_sites_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
mean_metastatic_sites_df_pivot.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
standard_error_metastatic_sites_df_pivot = standard_error_metastatic_sites_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
standard_error_metastatic_sites_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_metastatic_sites_df_pivot.index, mean_metastatic_sites_df_pivot["Capomulin"], yerr=standard_error_metastatic_sites_df_pivot["Capomulin"], marker="o", color="red", linestyle='--', label = "Capomulin")
plt.errorbar(mean_metastatic_sites_df_pivot.index, mean_metastatic_sites_df_pivot["Infubinol"], yerr=standard_error_metastatic_sites_df_pivot["Infubinol"], marker="^", color="blue", linestyle='--',label = "Infubinol")
plt.errorbar(mean_metastatic_sites_df_pivot.index, mean_metastatic_sites_df_pivot["Ketapril"], yerr=standard_error_metastatic_sites_df_pivot["Ketapril"], marker="s", color="green", linestyle='--', label = "Ketapril")            
plt.errorbar(mean_metastatic_sites_df_pivot.index, mean_metastatic_sites_df_pivot["Placebo"], yerr=standard_error_metastatic_sites_df_pivot["Placebo"], marker="d", color="black", linestyle='--', label = "Placebo")                         
 
# Set a Title and labels
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration(Days)")
plt.ylabel("Met. Sites")

# Set x & y limits of our scatter plot 
#plt.xlim(-1, 50)
#plt.ylim(30,80)

# Create a legend for the chart
plt.legend(loc="upper left")

#Set a grid on the plot
plt.grid()

plt.tight_layout()

# Save the Figure
plt.savefig("../Images/Metastatic Spread During Treatment.png")

# Show the figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_mice = merge_table.groupby(["Drug", "Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
count_mice_df = pd.DataFrame(count_mice)

# Rename "Mouse ID" to "Mouse Count"
count_mice_df = count_mice_df.rename(columns={"Mouse ID":"Mouse Count"})

# Reset DataFrame Index
count_mice_df.reset_index(inplace = True)

# Preview DataFrame
count_mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mice_df_pivot = count_mice_df.pivot(index="Timepoint", columns="Drug", values="Mouse Count")

# Preview that Reformatting worked
count_mice_df_pivot.head()
```

```python
## Create new columns and calculate survuval rate in %
count_mice_df_pivot["Capomulin_%"] = count_mice_df_pivot["Capomulin"]/count_mice_df_pivot["Capomulin"].iloc[0] * 100
count_mice_df_pivot["Infubinol_%"] = count_mice_df_pivot["Infubinol"]/count_mice_df_pivot["Infubinol"].iloc[0] * 100
count_mice_df_pivot["Ketapril_%"] = count_mice_df_pivot["Ketapril"]/count_mice_df_pivot["Ketapril"].iloc[0] * 100
count_mice_df_pivot["Placebo_%"] = count_mice_df_pivot["Placebo"]/count_mice_df_pivot["Placebo"].iloc[0] * 100
count_mice_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.plot(count_mice_df_pivot.index, count_mice_df_pivot["Capomulin_%"], marker="o", color="red", linestyle='--', label = "Capomulin")
plt.plot(count_mice_df_pivot.index, count_mice_df_pivot["Infubinol_%"], marker="^", color="blue", linestyle='--',label = "Infubinol")
plt.plot(count_mice_df_pivot.index, count_mice_df_pivot["Ketapril_%"], marker="s", color="green", linestyle='--', label = "Ketapril")            
plt.plot(count_mice_df_pivot.index, count_mice_df_pivot["Placebo_%"], marker="d", color="black", linestyle='--', label = "Placebo")                        
 
# Set a Title and labels
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Set x & y limits of our scatter plot 
#plt.xlim(-1, 50)
#plt.ylim(30,80)

# Create a legend for the chart
plt.legend(loc="best")

#Set a grid on the plot
plt.grid()

plt.tight_layout()

# Save the Figure
plt.savefig("../Images/Survival During Treatment.png")

# Show the figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = ((mean_tumor_volume_df_pivot.iloc[9] - mean_tumor_volume_df_pivot.iloc[0])/mean_tumor_volume_df_pivot.iloc[0]) * 100

# Display the data to confirm
percent_change.head(10)
```

# 

```python
# Store all Relevant Drug Names and Percent Changes into a List
drug = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

change = []

for item in drug:
    value = percent_change[item]
    change.append(value)
    
# Set a Title and labels
plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")

#Set a grid on the plot
plt.grid()

# Generate a bar graph
tumor_change = plt.bar(drug, change, alpha=0.80, align="center")

# Set the limits of the x axis
plt.xlim(-0.75, len(drug))

# Set the limits of the y axis
plt.ylim(-20, max(change)+5)

# Label the percentages of changes inside each bar displaying its height
def autolabel(rects):
    
    for rect in rects:
        height = rect.get_height()
        rect.set_color("red" if height >= 0 else "green")
        plt.text(rect.get_x() + rect.get_width()/2., 0.50*height,
                '%.0f%%' % float(height),
        ha='center', va='bottom')
autolabel(tumor_change)

plt.tight_layout()

# Save the Figure
plt.savefig("../Images/Tumor_Change_Over_Treatment.png")

# Show the figure
plt.show()
```
